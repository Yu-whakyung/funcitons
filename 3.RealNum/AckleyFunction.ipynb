```python
import numpy as np
import pandas as pd
from numpy import arange
from numpy import exp
from numpy import sqrt
from numpy import cos
from numpy import e
from numpy import pi
import sys
import time
from tqdm import notebook

def AckleyFunction (x):
    a = []
    col=len(x[0])
    row=(len(x))      
    for i in range(row):
        scd=0
        thr=0
        for j in range(col):
            scd += x[i][j]**2
            thr +=cos(2*pi*x[i][j])
            
        a.append(-20.0 * exp(-0.2 * sqrt(0.5 * (scd)))-exp(0.5 * (thr)) + e + 20)
        
    return a
```
