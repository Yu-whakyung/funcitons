```python
import numpy as np
import pandas as pd
import math

def RastriginFunction(x):
    row=len(x)
    col=len(x[0])
    y=[]
    for j in range(row):
        mid=[]
        final=0
        for i in range(col):
            mid = ((x[j][i])**2)-10*math.cos(2*math.pi*x[j][i])
            final = final+mid
        y.append(10*col+final)
    fi=np.array(y)
    return fi
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error 
from sklearn.svm import SVR
svm=SVR()

dataset=[1000,10000,30000]
dim = [5,10,15,20,30,40,50]

for i in dataset:
    for j in dim:
        csv_input1=pd.read_csv('./dataset/real/Random/{0}DiRanData_{1}.csv'.format(j,i))
        Rindata1 = csv_input1.values

        Routdata1=RastriginFunction(Rindata1)

        Rfidata1 = np.array(Routdata1)

        X_trainR1, X_testR1, y_trainR1, y_testR1 = train_test_split(Rindata1,Rfidata1,random_state=1234,train_size=0.7,test_size=0.3)

        ############### 일반 Normalize 랑 RMSE/maxInv, RMSE #########################

        print("### dim = {0} ###".format(j))
        print("### dataNum = {0} ###".format(i))


        ####Train

        #Rmodel1 = 
        Rtemp1 = svm.fit(X_trainR1,y_trainR1)
        R_pre1 = Rtemp1.predict(X_trainR1)

        ####Test
        #Z-score
        Rtemp_t1 = svm.fit(X_testR1,y_testR1) 
        R_pre_t1 = Rtemp_t1.predict(X_testR1)

        ErrArr = abs(R_pre_t1 - y_testR1)  #에러행렬
        ErrMean = np.mean(ErrArr)  #에러평균
        Sig = np.std(ErrArr)  #에러표준편차
        NoArr = (ErrArr - ErrMean)/Sig
        sumArr = sum(abs(NoArr))/len(NoArr)


        print(sumArr)

        RMSE_t1 = (mean_squared_error(y_testR1,R_pre_t1))**0.5  #RMSE(오차)
        print(RMSE_t1)
        Rnormal = RMSE_t1/j

        print(Rnormal)

```
