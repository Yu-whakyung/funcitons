```python
def schwefelFunction(x):  #100행 2열의 inputdata
    a = []
    #defalt =( x * 10000 ).astype(int)
    row = len(x)##/len(x[defalt])
    col=len(x[0])
    for n in range(row) :
        inpsum = 0
        for i in range(col):
            D1 = x[n][i]
            inpsum += D1*np.sin(np.sqrt(abs(D1)))
        final = 418.9829*col-inpsum
        a.append(final)

    return a
```
