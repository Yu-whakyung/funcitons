```python
import numpy as np
import pandas as pd
import RDFN

def SphereFunciton (x):
    a = []
    b=np.array(x)
    #defalt =( x * 10000 ).astype(int)
    row = len(b)##/len(x[defalt])
    col=int((np.size(b))/row)
          
    for y in range(row):
        sum=0
        for z in range(col):
            sum += (x[y][z])**2
        a.append(sum)
    return a
```
