```python
import numpy as np
import pandas as pd
import math
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error 
from sklearn.ensemble import RandomForestRegressor
```

```python


def GenerateBinaryNum (length,num):
    np.random.seed(1234)
    mGene = []
    fGene = []
    tempIn = []
    for j in range(num):
        tempIn=[]
        for i in range(length):
            sGene = (np.random.randint(1,100))%2
            tempIn.append(sGene)   
        mGene.append(tempIn)
    fGene=np.array(mGene)
    return fGene

def CalculateSchemaLength (gene):
    col = len(gene[0])
    n = (math.log(col))/(math.log(2))
    m = 2**(math.floor(((n-1)/2)+1))
        
    return m

def CalculateDelta(gene):
    col = len(gene[0])
    row = len(gene)
    dec = 0
    demal = []
    leng = len(gene)
    for j in range(row):
        for i in range(col):
            dec +=(gene[j][leng-i-1])*(2**i)
        demal.append(dec)
    Delta = math.floor(((((math.log(dec))/(math.log(2)))/2)+1)) 

    return Delta
```
