```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error 
from sklearn.ensemble import RandomForestRegressor


np.random.seed(1234)

def GenerateBinaryNum (length,num):
    mGene = []
    fGene = []
    tempIn = []
    for j in range(num):
        tempIn=[]
        for i in range(length):
            sGene = (np.random.randint(1,100))%2
            tempIn.append(sGene)   
        mGene.append(tempIn)
    fGene=np.array(mGene)
    bidf = pd.DataFrame(fGene)
    bidf.to_csv('{0}_{1}BinaryRanData.csv'.format(length,num),index=False)
    
    return fGene

def CalculateFitness(gene):
    fitness=0
    fitarr=[]
    col = len(gene[0])
    row = len(gene)
    for i in range(row):
        for j in range(col):
            if(gene[i][j]==1):
                fitness = fitness+1
        fitarr.append(fitness)
        fitness=0
    return fitarr


def binaryCaseCreate (length,num):
    wbase=[]
    wfinal=[]
    place= 2**(length)
    numcount=0
    while(numcount<num):
        for i in range(place):
            bi=format(i, "0{0}b".format(length))
            for j in range(length):
                wbase.append(bi[j])
            wnew = np.array(wbase)
            numcount = numcount+1
            print(wnew)
            if(numcount == num):
                wfinal = wnew.reshape(num,length)

    return wfinal

```
