```python
import numpy as np
import pandas as pd
import math
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error 
from sklearn.ensemble import RandomForestRegressor


def GenerateBinaryNum (length,num):
    np.random.seed(1234)
    mGene = []
    fGene = []
    tempIn = []
    for j in range(num):
        tempIn=[]
        for i in range(length):
            sGene = (np.random.randint(1,100))%2
            tempIn.append(sGene)   
        mGene.append(tempIn)
    fGene=np.array(mGene)
    return fGene

def CreateWBector (k):

#2^k+1 바이너리를 만들고 그에 상응하는 적합도벡터 w를 만든다
    wnum=2**(k+1)
    w= np.random.random(wnum)
    W= np.array(w)

    return W
    
#비트열 각 자리당 바이너리 계산
    
def checkdeci(Gene,K):
    bmach = []
    deci=0
    total=0
    men=[]
    k= int((math.log(2**(K+1))/math.log(2))-1)
    length = len(Gene[0])

    for u in range(len(Gene)):
        for j in range(length):
            for x in range(k+1):
                y=(j+x)%(length)
                c=Gene[u][y]
                deci += c*(2**(k-x))
            bmach.append(deci)
            deci=0
    fmach = np.array(bmach).reshape(len(Gene),length)
    
    col=len(fmach[0])
    row=len(fmach)

    #각 바이너리에 상응하는 적합도 벡터를 가져와서 평균적합도를 구한다.
    for y in range(row):
        for z in range(col):
            w=CreateWBector(K)
            indx = fmach[y][z]
            total += w[indx]
        men.append(total/col)
        total=0
    
    means = np.array(men)
    return means
```
