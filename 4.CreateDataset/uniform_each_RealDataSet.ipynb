```python
import numpy as np
import pandas as pd
from random import seed
from random import random

np.random.seed(1234)

def CreateRosenInput(dim,num):
    Rosen=[]
    fiRosen=[]
    for i in range(num):
        for j in range(dim):
            dim2_1000 = np.random.uniform(-2.048,2.048)
            Rosen.append(dim2_1000)
    fiRosen = np.array(Rosen)
    fiRosen = fiRosen.reshape(num,dim)
    Rosen=[]
    
    return fiRosen

def CreateRastriginInput(dim,num):
    Rastrigin=[]
    fiRastrigin=[]
    for i in range(num):
        for j in range(dim):
            dim2_1000 = np.random.uniform(-5.12,5.12)
            Rastrigin.append(dim2_1000)
    fiRastrigin = np.array(Rastrigin)
    fiRastrigin = fiRastrigin.reshape(num,dim)
    Rastrigin=[]
    
    return fiRastrigin

def CreateAckleyInput(dim,num):
    Ackley=[]
    fiAckley=[]
    for i in range(num):
        for j in range(dim):
            dim2_1000 = np.random.uniform(-32.768,32.768)
            Ackley.append(dim2_1000)
    fiAckley = np.array(Ackley)
    fiAckley = fiAckley.reshape(num,dim)
    Ackley=[]
    
    return fiAckley

def CreateSchwefelInput(dim,num):
    Schwefel=[]
    fiSchwefel=[]
    for i in range(num):
        for j in range(dim):
            dim2_1000 = np.random.uniform(-500,500)
            Schwefel.append(dim2_1000)
    fiSchwefel = np.array(Schwefel)
    fiSchwefel = fiSchwefel.reshape(num,dim)
    Schwefel=[]
    
    return fiSchwefel

def CreateSphereInput(dim,num):
    Sphere=[]
    fiSphere=[]
    for i in range(num):
        for j in range(dim):
            dim2_1000 = np.random.uniform(-5.12,5.12)
            Sphere.append(dim2_1000)
    fiSphere = np.array(Sphere)
    fiSphere = fiSphere.reshape(num,dim)
    Sphere=[]
    
    return fiSphere
```

```python
def CreateRosenInput_ver2(dim,num):
    Rosen=[]
    fiRosen=[]
    Rosen = (2.048+2.048)*np.random.random_sample(dim*num)-2.048
    fiRosen = np.array(Rosen)
    fiRosen = fiRosen.reshape(num,dim)
    return fiRosen
 
    
def CreateRastriginInput_ver2(dim,num):
    Rastrigin=[]
    fiRastrigin=[]
    Rastrigin = (5.12+5.12)*np.random.random_sample(dim*num)-5.12
    fiRastrigin = np.array(Rastrigin)
    fiRastrigin = fiRastrigin.reshape(num,dim)
    return fiRastrigin

def CreateAckleyInput_ver2(dim,num):
    Ackley=[]
    fiAckley=[]
    Ackley = (32.768+32.768)*np.random.random_sample(dim*num)-32.768
    fiAckley = np.array(Ackley)
    fiAckley = fiAckley.reshape(num,dim)
    return fiAckley

def CreateSchwefelInput_ver2(dim,num):
    Schwefel=[]
    fiSchwefel=[]
    Schwefel = (500+500)*np.random.random_sample(dim*num)-500
    fiSchwefel = np.array(Schwefel)
    fiSchwefel = fiSchwefel.reshape(num,dim)
    return fiSchwefel


def CreateSphereInput_ver2(dim,num):
    Sphere=[]
    fiSphere=[]
    Sphere = (5.12+5.12)*np.random.random_sample(dim*num)-5.12
    fiSphere = np.array(Sphere)
    fiSphere = fiSphere.reshape(num,dim)
    return fiSphere
```

```python
import matplotlib.pyplot as plt
a = CreateSchwefelInput_ver2(10,30000)
plt.hist(a)
plt.show()
```

```python
num_list = [30000]
dim_list = [2,4,8,16,20,30,40,50]

for i in num_list:
    for j in dim_list:
        fiRosen = CreateRosenInput_ver2(j,i)
        Rosen = pd.DataFrame(fiRosen)
        Rosen.to_csv('RanRosen_{0}_{1}.csv'.format(j,i),index=False)

        fiRastrigin = CreateRastriginInput_ver2(j,i)
        Rastrigin = pd.DataFrame(fiRastrigin)
        Rastrigin.to_csv('RanRastrigin_{0}_{1}.csv'.format(j,i),index=False)

        fiAckley = CreateAckleyInput_ver2(j,i)
        Ackley = pd.DataFrame(fiAckley)
        Ackley.to_csv('RanAckley_{0}_{1}.csv'.format(j,i),index=False)

        fiSchwefel = CreateSchwefelInput_ver2(j,i)
        Schwefel = pd.DataFrame(fiSchwefel)
        Schwefel.to_csv('RanSchwefel_{0}_{1}.csv'.format(j,i),index=False)

        fiSphere = CreateSphereInput_ver2(j,i)
        Sphere = pd.DataFrame(fiSphere)
        Sphere.to_csv('RanSphere_{0}_{1}.csv'.format(j,i),index=False)
```
