```python
import numpy as np
import pandas as pd
from random import seed
from random import random
import math
```

```python
def getPermuCount(arr):
    row = len(arr)
    permu_num = []
    for j in range(row):
        col = len(arr[j])
        num = my_factorial(col)
        permu_num.append(num)
        num=0
    return permu_num

def getInvCount(arr):
    inv_count_row = []
    col = len(arr[0])
    row = len(arr)
    for k in range(row):
        inv_count = 0
        for i in range(col-1):
            for j in range(i+1,col):
                if (arr[k][i] > arr[k][j]):
                    inv_count += 1
        inv_count_row.append(inv_count)
    return inv_count_row

def MakeZeroInv(arr):
    for i in range(len(arr)-1):
        for j in range(i+1,len(arr)):
            if((arr[i])>(arr[j])):
                temp=arr[i]
                arr[i]=arr[j]
                arr[j]=temp
    return arr
        
        

def my_factorial(n): 
    if(n > 1): 
        return n * my_factorial(n - 1) 
    else: 
        return 1
    
```

```python
def CreateRangePermuArr_Ver22 (dim, num):

    count= 0 
    maxInvNum = (int)(dim*(dim-1)/2)
    newa=[]
    permu=[]
    fi=[]
    loop = math.ceil(num/maxInvNum)
    
    for i in range(loop):
        seed(i)
        permu=[]
        for n in range(dim):
            sed = random()
            new = sed
            permu.append(new)
        permu = MakeZeroInv(permu)
        newa.extend(permu)

        count = count +1
        for j in range(dim):
            k=dim-j-1
            for m in range(k):
                if(count<num):
                    temp = permu[m]
                    permu[m] = permu[m+1]
                    permu[m+1]=temp
                    newa.extend(permu)

                    count = count+1
                if(count==num):
                    fi = np.array(newa).reshape(num,dim)
    
    result_df = pd.DataFrame(fi)   #데이터 프레임 형식으로 만들기
    df_shuffled = result_df.sample(frac=1).reset_index(drop=True)  #섞기
    return df_shuffled
            
```
