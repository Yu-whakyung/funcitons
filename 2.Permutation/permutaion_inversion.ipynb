```python
import numpy as np
import pandas as pd
import RDFN
import CreateRangePermuArr
import FreckPernuArr


def getInvCount(arr):
    inv_count_row = []
    col = len(arr[0])
    row = len(arr)
    for k in range(row):
        inv_count = 0
        for i in range(col-1):
            for j in range(i+1,col):
                if (arr[k][i] > arr[k][j]):
                    inv_count += 1
        inv_count_row.append(inv_count)
    return inv_count_row
  
def getPermuCount(arr):
    row = len(arr)
    permu_num = []
    for j in range(row):
        col = len(arr[j])
        num = my_factorial(col)
        permu_num.append(num)
        num=0
    return permu_num

def my_factorial(n): 
    if(n > 1): 
        return n * my_factorial(n - 1) 
    else: 
        return 1
    
def maxInvNum(arr):
    col = len(arr[0])
    maxnum=(col*(col-1))/2
    return maxnum
    
    
```

```python
################## Shuffled 데이터셋 ##################

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error 
from sklearn.svm import SVR
svm=SVR()

dataset=[30000]
dim = [5,10,15,20,30,40,50]

for i in dataset:
    for j in dim:
        csv_input1=pd.read_csv('./dataset/real/Shuffle/Shuffle_{0}_{1}.csv'.format(j,i))
        Rindata1 = csv_input1.values

        Routdata1=getInvCount(Rindata1)

        Rfidata1 = np.array(Routdata1)

        X_trainR1, X_testR1, y_trainR1, y_testR1 = train_test_split(Rindata1,Rfidata1,random_state=1234,train_size=0.7,test_size=0.3)

        ############### 일반 Normalize 랑 RMSE/maxInv, RMSE #########################

        print("----------------### dim = {0} ###-------------------".format(j))
        print("----------------### dataNum = {0} ###----------------".format(i))


        ####Train

        #Rmodel1 = 
        Rtemp1 = svm.fit(X_trainR1,y_trainR1)
        R_pre1 = Rtemp1.predict(X_trainR1)

        ####Test
        R_pre_t1 = Rtemp1.predict(X_testR1)
        
        #Z-score        
        ErrArr = abs(R_pre_t1 - y_testR1)  #에러행렬
        ErrMean = np.mean(ErrArr)  #에러평균
        Sig = np.std(ErrArr)  #에러표준편차
        NoArr = (ErrArr - ErrMean)/Sig
        sumArr = sum(abs(NoArr))/len(NoArr)


        print(sumArr)

        RMSE_t1 = (mean_squared_error(y_testR1,R_pre_t1))**0.5  #RMSE(오차)
        print(RMSE_t1)
        Rnormal = RMSE_t1/maxInvNum(Rindata1)

        print(Rnormal)
```
